```python
data_nyu = pd.read_csv('sdss_nyu.csv.bz2')
data_fs2 = pd.read_csv('sdss_fs2.csv.bz2')
```

```python
z_min_sdss_fit = 0.15 # minimum redshift used for fitting number density in FS2 to observations
dr_fs2_sdss = -0.15 # shift in absolute r-band magnitude in FS2, chosen to match observed number density above redshift z_min_sdss_fit
```

# compare number densities

```python
#==== NYU VAGC DR7 =====
A_sdssNYU = 7966. #see http://sdss.physics.nyu.edu/vagc/
Ng_sdssNYU = len(data_nyu[data_nyu.z>z_min_sdss_fit])
ng_sdssNYU = Ng_sdssNYU / A_sdssNYU

#==== Flagship 2 =====
A_oct = 5156.62 # 1 octant in deg^2
Ng_fs2_sdss = len(data_fs2[data_fs2.observed_redshift_gal>z_min_sdss_fit])
ng_fs2_sdss = Ng_fs2_sdss / A_oct

print (A_sdssNYU, Ng_sdssNYU, ng_sdssNYU)
print (A_oct, Ng_fs2_sdss, ng_fs2_sdss)
```

# plot number density vs redshift

```python
fig, ax = plt.subplots(1,1,figsize=(5,5))

fontsize=18

plt.rc('font', family='serif')
plt.rc('xtick', labelsize=fontsize)
plt.rc('ytick', labelsize=fontsize)
plt.rc('text', usetex=True)

z_bins = np.linspace(0.08,0.35,100)

ax.set_title('SDSS main', fontsize=fontsize)

#redshift bins
bins = bins=np.linspace(0,0.4,80)

#plot histogram for NYU sample
hist, bins_edges = np.histogram(data_nyu.z, bins=bins)
bin_centers = (bins_edges[1:] + bins_edges[:-1])*0.5
ax.plot(bin_centers, hist/A_sdssNYU, label='SDSS DR7', color='k', ls='-', lw=2)

#plot histogram for FS2 sample
hist, bins_edges = np.histogram(data_fs2.observed_redshift_gal, bins=bins)
bin_centers = (bins_edges[1:] + bins_edges[:-1])*0.5
ax.plot(bin_centers, hist/A_oct, label='FS 2', color='r', ls='--', lw=2)

ax.set_xlim(0.06,0.32)
ax.set_xlabel('redshift', fontsize=fontsize)
ax.set_ylabel('$n_{gal}$', fontsize=fontsize)
ax.axvline(x=z_min_sdss_fit, linestyle='--', c='black')

ax.legend(loc = 'upper right', fontsize=fontsize)
    
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1,2,figsize=(8,4), sharey=True, sharex=True)

fontsize=18

plt.rc('font', family='serif')
plt.rc('xtick', labelsize=fontsize)
plt.rc('ytick', labelsize=fontsize)
plt.rc('text', usetex=True)

m_bins = np.linspace(15,18.2,200)
z_bins = np.linspace(0.05,0.3,200)

ax[0].set_title('SDSS DR7', fontsize=fontsize)
ax[1].set_title('Flagship 2', fontsize=fontsize)

Nmin, Nmax = 0,1000

A_sdss= 7966. #see http://sdss.physics.nyu.edu/vagc/
A_oct = 5156.62 # 1 octant in deg^2
fA= A_sdss / A_oct

ax[0].hist2d(data_nyu.z, data_nyu.mr_nodust, bins=[z_bins,m_bins], cmin=Nmin*fA, cmax=Nmax*fA, cmap='Greys')
ax[1].hist2d(data_fs2.observed_redshift_gal, data_fs2.sdss_r_app_mag, 
             bins=[z_bins,m_bins], cmin=Nmin, cmax=Nmax, cmap='Greys')

for i in range(0,2):
    ax[i].axhline(y=17.77, c='r', label='SDSS main cut')

for i in range(0,2):
    ax[i].set_xlabel('$z$', fontsize=fontsize)
    ax[i].axvline(x=z_min_sdss_fit, linestyle='--', c='k', linewidth=2)


ax[0].set_ylabel('$m_r$', fontsize=fontsize)

ax[1].axhline(y=17.77+dr_fs2_sdss, c='r', linestyle='--', label='modified cut')

ax[1].legend(loc='lower right', fontsize=fontsize, framealpha=1.0)
plt.tight_layout()
#plt.savefig('mr-z_fs2_sdss.png', dpi=400,bbox_inches='tight', pad_inches = 0.1)
plt.show()
```

```python
fig, ax = plt.subplots(3,1,figsize=(5,10), sharex=True)

fname = 'mr-z_fs2_sdss.pdf'

fontsize=18

plt.rc('font', family='serif')
plt.rc('xtick', labelsize=fontsize)
plt.rc('ytick', labelsize=fontsize)
plt.rc('text', usetex=True)

plt.subplots_adjust(hspace=0.1)

m_bins = np.linspace(15,18.2,200)
z_bins = np.linspace(0.05,0.3,200)

label_nyu = 'NYU-VAGC'
label_fs2 = 'FS 2'

ax[0].text(0.33, 16, label_nyu, {'color': 'k', 'fontsize': fontsize}, rotation=-90)
ax[1].text(0.33, 16, label_fs2, {'color': 'k', 'fontsize': fontsize}, rotation=-90)

Nmin, Nmax = 0,1000

fA= A_sdssNYU / A_oct

#========= magntude-redshift distribution ===========
ax[0].hist2d(data_nyu.z, data_nyu.mr_nodust,
             bins=[z_bins,m_bins], cmin=Nmin*fA, cmax=Nmax*fA, cmap='Greys')
ax[1].hist2d(data_fs2.observed_redshift_gal, data_fs2.sdss_r_app_mag, 
             bins=[z_bins,m_bins], cmin=Nmin, cmax=Nmax, cmap='Greys')

for i in range(0,2):
    ax[i].axhline(y=17.77, c='r', label='SDSS main cut')
    ax[i].set_ylabel('$m_r$', fontsize=fontsize)
    ax[i].set_ylim(15,18.2)

ax[1].axhline(y=17.77+dr_fs2_sdss, c='r', linestyle='--', label='modified cut')


#========= redshift distribution ===========

z_bins = np.linspace(0.08,0.35,100)

#redshift bins
bins = bins=np.linspace(0,0.4,80)

#plot histogram for NYU sample
hist, bins_edges = np.histogram(data_nyu.z, bins=bins)
bin_centers = (bins_edges[1:] + bins_edges[:-1])*0.5
ax[2].plot(bin_centers, hist/A_sdssNYU, label=label_nyu, color='k', ls='-', lw=2)

#plot histogram for FS2 sample
hist, bins_edges = np.histogram(data_fs2.observed_redshift_gal, bins=bins)
bin_centers = (bins_edges[1:] + bins_edges[:-1])*0.5
ax[2].plot(bin_centers, hist/A_oct, label=label_fs2, color='k', ls='--', lw=2)

ax[2].set_xlim(0.06,0.32)
ax[2].set_ylabel('$N_{gal}/deg^2$', fontsize=fontsize)

for i in range(0,3):
    ax[i].axvline(x=z_min_sdss_fit, linestyle='--', c='k', linewidth=2)
ax[-1].set_xlabel('$z$', fontsize=fontsize)




ax[1].legend(loc='lower right', fontsize=fontsize, framealpha=0.0)
ax[2].legend(loc='upper right', fontsize=fontsize)



#plt.tight_layout()
plt.savefig(fname, dpi=400,bbox_inches='tight', pad_inches = 0.1)
plt.show()
```

```python
fig, ax = plt.subplots(1,1,figsize=(5,5))

fname='ur_fs2_sdss.pdf'

fontsize=18

plt.rc('font', family='serif')
plt.rc('xtick', labelsize=fontsize)
plt.rc('ytick', labelsize=fontsize)
plt.rc('text', usetex=True)

gr_bins = np.linspace(0.0,1.2,100)

gr_nyu = data_nyu.Mg00 - data_nyu.Mr00
gr_fs2 = data_fs2.gr_restframe


gr_cut_fs2 = 0.61
gr_cut_nyu = 0.62
gr_cut_J19 = 0.66

ax.axvline(x=gr_cut_fs2,c='r', ls='--', lw=4)
#ax.axvline(x=gr_cut_nyu,c='k', ls='--', lw=2)
#ax.axvline(x=gr_cut_J19,c='b', ls='--', lw=2)

ax.hist(gr_nyu, bins=gr_bins, histtype='step', density=True, color='k', lw=1, label='SDSS')
ax.hist(gr_fs2, bins=gr_bins, histtype='step', density=True, color='r', lw=1, label='FS2')

ax.set_xlabel('$M_u-M_r$',fontsize=fontsize)
ax.set_ylabel('$pdf$',fontsize=fontsize)

ax.legend(fontsize=fontsize, loc='upper left')
plt.savefig(fname, dpi=400,bbox_inches='tight', pad_inches = 0.1)

plt.show()
```
